

The features of the datasets are not normally distributed, so there is a need to make them normally distributed. We will make use of qq plots

Before, making the data noprmally distributed let's divide the whole dataset into X and y


```python
x=data.drop("concrete_compressive_strength",axis=1)
y=data[["concrete_compressive_strength"]]
```

```python
x
```

```python
y
```



Transforming the data into normal distribution using box-cox transformer


```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer()
feature_normalization=pt.fit_transform(x)
```

```python
feature_normalised=pd.DataFrame(feature_normalization,columns=x.columns)

from scipy import stats
```

```python
for col in x.columns:
    plt.figure(figsize=(15,5))
    plt.subplot(2,2,1)
    stats.probplot(x=x[col],dist="norm", plot=plt)
    plt.title(col)
    plt.figure(figsize=(10,5))
    plt.subplot(2,2,2)
    stats.probplot(x=feature_normalised[col],dist="norm", plot=plt)
    plt.title(label=col + "After Transformation")
    plt.subplot(2,2,3)
    sns.kdeplot(x=x[col])
    plt.subplot(2,2,4)
    sns.kdeplot(x=feature_normalised[col])
    plt.show()
    
    
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=40)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
print('r2_score_test:',lr.score(x_test,y_test))
print('r2_score_train:',lr.score(x_train,y_train))
```

```python
y_pred=lr.predict(x_test)
lr_pred_df=pd.DataFrame(np.c_[y_test,y_pred],columns=['Actual compressive strength','Predicted compressive strength by LR'])
lr_pred_df
```

```python
rf= RandomForestRegressor()
rf.fit(x_train,y_train)
print("r2_score_test:",rf.score(x_test,y_test))
print("r2_score_train:",rf.score(x_train,y_train))
```

```python
y_pred=rf.predict(x_test)
rf_pred_df=pd.DataFrame(np.c_[y_test,y_pred],columns=['Actual compressive strength','Predicted compressive strength by LR'])
rf_pred_df
```

```python
from sklearn.ensemble import AdaBoostRegressor
ad_rg=AdaBoostRegressor()
ad_rg.fit(x_train,y_train)
print("r2_score_test:",ad_rg.score(x_test,y_test))
print("r2_score_train:",ad_rg.score(x_train,y_train))
```

```python
from xgboost import XGBRegressor
xgbr=XGBRegressor()
xgbr.fit(x_train,y_train)
print(xgbr.score(x_train,y_train))
print(xgbr.score(x_test,y_test))
```

```python
params = {'booster': ['dart', 'gbtree'],'learning_rate': [0.025, 0.05], 'max_depth': [3, 5, 6],'n_estimators': [100, 200, 300],'subsample': [0.6, 0.8]}
from sklearn.model_selection import GridSearchCV
grid_search=GridSearchCV(xgbr,params, cv = 2, n_jobs = -1,verbose=2)
grid_search.fit(x_train, y_train)
```

```python
 grid_search.best_params_
```

```python
from xgboost import XGBRegressor
xgbr_gs=XGBRegressor(booster= 'gbtree',learning_rate= 0.05,max_depth= 5,n_estimators= 300,subsample= 0.6)
xgbr_gs.fit(x_train,y_train)
print("r2_score_train:",xgbr_gs.score(x_train,y_train))
print("r2_score_test:",xgbr_gs.score(x_test,y_test))
```

```python
y_pred=xgbr_gs.predict(x_test)
xgbr_gs_pred_df=pd.DataFrame(np.c_[y_test,y_pred],columns=['Actual compressive strength','Predicted compressive strength by LR'])
xgbr_gs_pred_df
```
